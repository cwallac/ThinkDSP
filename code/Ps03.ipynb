1. Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
import numpy as np
import thinkdsp 
import thinkplot


bitFile = open("bitcoin.csv")
bitString = bitFile.read()
#bitList = map(float,bitString.split("\n"))
bitList = bitString.split("\n")
del bitList[-1] #Last entry is space will cause error
bitList = map(float,bitList)
prices = np.array(bitList)

bitWave = thinkdsp.Wave(prices,1) # Wave with 1 entry per day (TIME UNIT'S ARE PER DAY)
spectrum = bitWave.make_spectrum()
#spectrum.low_pass(1.0/7) #Removes trends that are less than 1 week 
#spectrum.high_pass(0.001)
spectrum.plot_power()
print "Beta value is :" ,spectrum.estimate_slope()[0]
thinkplot.config(low=1,alpha=0.5,xscale='log',yscale='log')
```

The power spectrum of bitcoin prices is almost brownian noise. It has a beta value of 1.79, so it is technically still pink noise, but very close to 2 which would qualify it as brownian noise.

2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
lags, corrs = autocorr(bitWave)
thinkplot.plot(lags, corrs)
thinkplot.config(xlabel='lag (index)', ylabel='correlation', ylim=[-1, 1])
```

There is linear decay in the correlation of bitcoin prices, this intuitevely makes sense. While correlation is below .5 after around 200 days there are bumps in correlation at around 450 days and 700 days, indicating that there might be some form of slightly periodic behavior who's period is decreasing with time.

4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

```python
import dct
import scipy.fftpack
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** np.arange(4, 12)
results = []
results2 = []
results3 = []
for N in ns:
    ts = (0.5 + np.arange(N)) / N
    freqs = (0.5 + np.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)
    result2 = %timeit -o dct.analyze2(ys, freqs, ts)
    results2.append(result2)
    result3 = %timeit -o scipy.fftpack.dct(ys)
    results3.append(result3)

bests = [result.best for result in results]
bests2 = [result.best for result in results2]
bests3 = [result.best for result in results3]

thinkplot.plot(ns, bests,label="Analyze 1")
thinkplot.plot(ns, bests2,label="Analyze 2")
thinkplot.plot(ns, bests3,label="scipy.fftpack.dct")
thinkplot.config(xscale='log', yscale='log',legend=True,loc="upper left")
```

```python
x = np.log(ns)
y1 = np.log(bests)
y2 = np.log(bests2)
y3 = np.log(bests3)
t1 = scipy.stats.linregress(x,y1)
t2 = scipy.stats.linregress(x,y2)
t3 = scipy.stats.linregress(x,y3)
print t1[0], t2[0], t3[0]
```

The analyze 1 and analyze 2 methods run with very similar orders of growth. O(n^2) roughly. HOwever, the dct runs with an order growth of less than one. Which means it must run in logarithmic time. So it is considerably better than analyze1 or 2
