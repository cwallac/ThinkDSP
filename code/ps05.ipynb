2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

So, we know that the effect of a window can be seen by multiplying the DFT of the signal with the DFT of the window. IF we want our filter to be ideal we need one where the amplitude of all components below the cut off frequency (assuming this is a high pass filter) is 1 and all the amplitudes of frequencies above that catoff frequency are zero. Using the ThinkDSP libary we can form that spectrum with the code below.

```python
futureHs = []
for i in range(5513):
    if i <= 440:
        futureHs.append(-1j)
    else:
        futureHs.append(0)

idealFilterSpect = thinkdsp.Spectrum(futureHs,11025)    
```

```python
idealFilterSpect.plot()
```

Spectrum looks ideal for a high pass filter with cutoff of 440

```python
idealFilterSpect.make_wave().plot()
```

This is what the window with that spectrum looks like, so our ideal filter window looks like this. Below is a zoomed in view of the beginning of the high pass filter. 

```python
idealFilterSpect.make_wave().segment(0,.05).plot()
```

We can repeat this process for a low pass filter and we will show the results below

```python
futureHs = []
for i in range(5513):
    if i >= 440:
        futureHs.append(-1j)
    else:
        futureHs.append(0)

idealFilterSpect = thinkdsp.Spectrum(futureHs,11025)   
```

3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
import numpy

signal = thinkdsp.SquareSignal(freq=440)
wave = signal.make_wave(duration=1, framerate=44100)
segment = wave.segment(duration=1)
window = numpy.ones(11)
window /= sum(window)
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

N = len(segment)
padded = zero_pad(window, N)


wave = thinkdsp.Wave(padded,11025)
tesAn = wave.make_spectrum().angles()
thinkplot.plot(tesAn)
```

The phase offset increases linearly with frequency, the rapid increases are an articfact of the modulo nature of offset angles.

```python
import scipy.signal
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
padded = zero_pad(gaussian, N)
wave = thinkdsp.Wave(padded,11025)
tesAn = wave.make_spectrum().angles()
thinkplot.plot(tesAn)
```

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

```python
window = numpy.asarray([-.5,2,-1.5])
padded = zero_pad(window,len(wave))
dft_window = numpy.fft.rfft(padded)
thinkplot.plot(abs(dft_window), color='0.7', label='Second order forard approx filter')


thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='amplitude ratio',
                 xlim=[0, 22050], 
                 legend=False)
```

The second order appriximation looks lamost like a line, it does have curves that would be associated with a second order approximation, but they are very close to a  line where the amplitude is proportional to f

5) The gunshot recording I used in chap09.ipynb was clipped, which means that the loudest part gets cut off, which makes the rest of the sound seem louder than it should be, which exaggerates the reverb of the space.

But it looks like there is a better version of this recording at http://www.freesound.org/people/Kleeb/sounds/180960/

Use this recording to replace the one I used and see if it tones down the reverb a little.

Also, apply the same effect to the saxophone recording.

```python
response = thinkdsp.read_wave('GunShot.wav')
response = response.segment(start=0.20, duration=5.0)
response.normalize()
response.plot()
thinkplot.config(ylim=[-1.05, 1.05])
transfer = response.make_spectrum()
```

```python
violin = thinkdsp.read_wave('92002__jcveliz__violin-origional.wav')
violin.ys = violin.ys[:len(response)]
violin.normalize()
spectrum = violin.make_spectrum()
violin.make_audio()
```

```python
output = (spectrum * transfer).make_wave()
output.normalize()
output.make_audio()
```

The reverb was definietely toned downa s compared to the original gunshot clip.

```python
sax = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
sax.ys = sax.ys[:len(response)]
sax.normalize()
spectrum = sax.make_spectrum()
sax.make_audio()
```

```python
output = (spectrum * transfer).make_wave()
output.normalize()
output.make_audio()
```

The saxophone sounds much more distant than the violin recording
