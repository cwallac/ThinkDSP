Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal. 

```python
class SawtoothSignal(thinkdsp.Sinusoid):

    def evaluate(self,ts):
        period = 1.0/self.freq
        normalized = np.mod(ts+self.offset,period)/period #modulo and normalize to get from 0 to 1
        return 2*normalized-1 # Value from -1 to 1 

saw = SawtoothSignal(200)

duration = saw.period*3
sawseg = saw.make_wave(duration,framerate=10000)

sawseg.plot()
        
        
    
```
